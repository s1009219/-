```python
btree = [None] * 1024
btree[1] = 'A'
btree[2] = 'B'
btree[3] = 'C'
btree[4] = 'D'
btree[5] = 'E'  #有沒有發現幾乎只是照順序填下來而已，沒得就空開
btree[7] = 'F'
```

```python
def preorder(p):
  if btree[p] != None:
    print(btree[p], end=' ')
    preorder(2*p)
    preorder(2*p+1)

def inorder(p):
  if btree[p] != None:
    inorder(2*p)
    print(btree[p], end=' ')
    inorder(2*p+1)

def postorder(p):
  if btree[p] != None:
    postorder(2*p)
    postorder(2*p+1)
    print(btree[p], end=' ')
```

```python
#用鏈結串列的方式來建立，可以節省很多記憶體(尤其是未平衡前的二元樹)
class btree:
  def __init__(self, value):
    self.data = value
    self.left = self.right = None

  def setLeft(self, left):
    self.left = left

  def setRight(self, right):
    self.right = right
```

```python
p1 = btree('A')
p2 = btree('B')
p3 = btree('C')
p4 = btree('D')
p5 = btree('E')
p6 = btree('F')
```

```python
root = p1
p1.setLeft(p2)
p1.setRight(p3)
p2.setLeft(p4)
p2.setRight(p5)
p3.setRight(p6)
```

```python
def linkPreOrder(p):
  if p != None:
    print(p.data, end=' ')
    linkPreOrder(p.left)
    linkPreOrder(p.right)

def linkInOrder(p):
  if p != None:
    linkInOrder(p.left)
    print(p.data, end=' ')
    linkInOrder(p.right)

def linkPostOrder(p):
  if p != None:
    linkPostOrder(p.left)
    linkPostOrder(p.right)
    print(p.data, end=' ')
```

```python
linkPreOrder(root)
```
